```python
import numpy as np
from keras.models import Sequential, Model
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Input, Embedding
from sklearn.model_selection import train_test_split
from keras.layers.merge import Concatenate
```

```python
data = open("PatientDetect.txt", "r")
tdata = np.empty((0,24),dtype=float)
tlabel = np.empty((0,3),dtype=float)
train_images = []
for d in data.readlines():
    tdata = np.array([float(t) for t in d.split(",")[:24]]).reshape((1,24))
    labelNum = float(d.split(",")[24].strip("\n"))
    labelV = np.zeros((3,1))
    labelV[int(labelNum)] = 1
    tlabel = np.append(tlabel, labelV.reshape(1,3), axis=0)
    train_images.append(tdata.reshape(2,4,3))
train_images = np.array(train_images)
print(train_images.shape)
print(tlabel.shape)
```

```python
max0 = 0
max1 = 0
for i in range(train_images.shape[0]):
    temp = max(np.max((train_images[i][0])), np.max((train_images[i][0])))
    if(max0 < temp):
        max0 =  temp
    temp = max(np.max((train_images[i][1])), np.max((train_images[i][1])))
    if(max1 < temp):
        max1 =  temp
print(max0)
print(max1)
```

```python
train_images = (train_images / max0)
```

```python
train_images = np.moveaxis(train_images, [1, 2, 3], [3, 1, 2])
print(train_images.shape)
```

```python
num_filters = 2
filter_size = 2

model = Sequential([
  Conv2D(num_filters, filter_size, input_shape=(4, 3, 2)),
  Flatten(),
  Dense(12, activation='linear'),
    Dense(6, activation='linear'),
    Dense(3, activation='softmax'),
])
```

```python
model.compile(
  'adam',
  loss='binary_crossentropy',
  metrics=['accuracy'],
)
```

```python
x_train, x_test, y_train, y_test = train_test_split(train_images, tlabel, test_size=0.05)
```

```python
model.fit(
    x_train,
    y_train,
    epochs=500,
    validation_data=(x_test, y_test),)
```

```python
model.save_weights('body.h5')
```

```python
model.load_weights('body.h5')
```

```python
data = open("PatientWeight.txt", "r")
tdata = []
tlabel = np.empty((0,1),dtype=float)
bodyOutput = []
train_images = []
for d in data.readlines():
    temp = np.array([float(t) for t in d.split(",")[:25]]).reshape((1,25))
    tdata = temp[0][:24]
    bodyOutput.append(temp[0][24])
    tlabel = np.append(tlabel, np.array([float(d.split(",")[25].strip("\n"))]).reshape(1,1) ,axis=0)
    train_images.append(np.array(tdata).reshape(2,4,3))
train_images = np.array(train_images)
bodyOutput = np.array(bodyOutput)
bodyOutput = bodyOutput.reshape(bodyOutput.shape[0],1)
print(train_images.shape)
print(bodyOutput.shape)
print(tlabel.shape)
```

```python
max0 = 0
max1 = 0
for i in range(train_images.shape[0]):
    temp = max(np.max((train_images[i][0])), np.max((train_images[i][0])))
    if(max0 < temp):
        max0 =  temp
    temp = max(np.max((train_images[i][1])), np.max((train_images[i][1])))
    if(max1 < temp):
        max1 =  temp
print(max0)
print(max1)
```

```python
train_images = (train_images / max0)
```

```python
train_images = np.moveaxis(train_images, [1, 2, 3], [3, 1, 2])
print(train_images.shape)
```

```python
num_filters = 2
filter_size = 2

sequence_input = Input(batch_shape=train_images.shape, dtype=float, name='text_input')
#model = Embedding(input_dim=5, output_dim=1)(sequence_input)
model = (Conv2D(filters=num_filters, kernel_size=filter_size, input_shape=(4, 3, 2)))(sequence_input)
model = (Flatten())(model)
af_input = Input(batch_shape=(bodyOutput.shape), name='af_input')
model = Concatenate()([model, af_input])
model = (Dense(12, activation='linear'))(model)
model = (Dense(6, activation='linear'))(model)
model = (Dense(3, activation='softmax'))(model)
```

```python
model = Model(inputs=[sequence_input, af_input], outputs=model)
model.compile(
  'adam',
  loss='binary_crossentropy',
  metrics=['accuracy'],
)
```

```python
x_train, x_test, y_train, y_test = train_test_split(train_images, tlabel, test_size=0.05)
```

```python
model.fit(
    x_train,
    y_train,
    epochs=500,
    validation_data=(x_test, y_test),)
```

```python
model.save_weights('weight.h5')
```

```python
model.load_weights('weight.h5')
```
